### 2.Aggregate DataFrame df by extracting the month and year from the 'Date' column. Calculate the collective revenue for each unique combination of month and year.



```python
df[['Year', 'Month', 'Day']] = df['Date'].str.split('-', expand=True) #split the date in the form of date year month

df[['Year', 'Month', 'Day']] = df[['Year', 'Month', 'Day']].apply(pd.to_numeric) # convert str in to numeric

df.groupby(['Year', 'Month'])['Revenue'].sum()
```

### 3.Group the DataFrame df by the 'Date' column and find the date with the highest total revenue and the date with the highest total quantity sold.



```python
df.groupby(['Date'])[['Revenue','Quantity']].sum()
```

### 5. Group the DataFrame df by the 'Product' column and add a new column 'Revenue Percentage' that shows the percentage of each product's revenue relative to the total revenue for that product.


```python
product = df.groupby('Product')['Revenue'].transform('sum')

#transform('sum') calculates the 'Revenue Percentage' for each row by dividing the individual revenue by the total revenue for that product. 

df['Revenue Percentage'] = (df['Revenue'] / product) * 100
```
